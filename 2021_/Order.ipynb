```python
from Kiwoom_API import Kiwoom
import pandas as pd
import sys
from PyQt5.QtCore import *
from PyQt5.QtGui import *
from PyQt5.QtWidgets import *
import re 

if_first = True

def Send_order(account, order_type, code, num, price):
    order_type_lookup = {'신규매수': 1, '신규매도': 2, '매수취소': 3, '매도취소': 4}
    hoga_lookup = "03"
    kiwoom.send_order("send_order_req", "0101", account, order_type_lookup[order_type], code, num, price, hoga_lookup, "")

signal_df = pd.read_csv("ochlv_csv/시그널/Signal.csv",index_col=0, low_memory=False)
signal_df.index = pd.DatetimeIndex(signal_df.index)

last_rebalance_time = "20211104"
rebalance_time = "20211111"

last_bucket = signal_df.loc[last_rebalance_time][signal_df.loc[last_rebalance_time] =="Buy"].index
current_bucket = signal_df.loc[rebalance_time][signal_df.loc[rebalance_time] =="Buy"].index

hold_list = list(set(current_bucket).intersection(set(last_bucket)))
sell_list = list(set(last_bucket) - set(hold_list))
buy_list = list(set(current_bucket) - set(hold_list))

app = QApplication(sys.argv)
kiwoom = Kiwoom()
kiwoom.comm_connect()

account = kiwoom.get_login_info("ACCNO")
print(account)
account = account.split(";")[0]

kiwoom.set_input_value("계좌번호", account)
kiwoom.comm_rq_data("opw00018_req", "opw00018", 0, "2000")

kiwoom.set_input_value("계좌번호", account)
kiwoom.comm_rq_data("opw00001_req", "opw00001", 0, "2000")

balance_sheet = pd.DataFrame(kiwoom.opw00018_output['multi'],columns=["종목명","보유량","매입가","현재가","평가손익","수익률","종목번호"])
balance_sheet["종목번호"] = balance_sheet["종목번호"].apply(lambda x: re.sub(r'[A-Z]+', '', x,re.I))
print(balance_sheet)

for i in buy_list:
    if i.replace("'","") not in balance_sheet["종목번호"]:
        Send_order(account, "신규매수", i.replace("'",""), 1, "")
for i in sell_list:
    if  i.replace("'","") in balance_sheet["종목번호"]:
        Send_order(account, "신규매도", i.replace("'",""), 1, "")
    else:
        code = i.replace("'","")
        print(f"현재 보유하고 있지 않은 종목입니다. {code}")

if if_first == True:
    for i in hold_list:
        if i.replace("'","") not in balance_sheet["종목번호"]:
            Send_order(account, "신규매수", i.replace("'",""), 1, "")
```
